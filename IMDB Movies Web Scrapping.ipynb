```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("https://www.imdb.com/chart/top/")
page = r.text
soup = bs(page,'html.parser')
```

```python
# Initializing variables
movies_list = []
years = []
rating = []

# fetching data from website through HTML parser
movies_class = soup.select('td.titleColumn')
rating_all = soup.find_all('td', class_ = 'ratingColumn imdbRating')

for movie in movies_class:
    moviename = movie.find("a").get_text()
    saal = movie.find("span").get_text()
    years.append(saal)
    movies_list.append(moviename)
    
for r in rating_all:
    ratingsss = r.find("strong").get_text()
    ratingsss = float(ratingsss)
    rating.append(ratingsss)
    
```
